Arrays in NumPy are of fixed size and homogeneous in nature. NumPy arrays provide N-dimensional array objects that are used in linear algebra, Fourier Transformation, and random number capabilities. These array objects are much faster and more efficient than the Python Lists.

## 1. Creating One-Dimensional Arrays
NumPy one-dimensional arrays are linear arrays created from Python lists, tuples, or using the `fromiter()` function.

### Methods
- **From Python List**: `np.array([1, 2, 3, 4, 5])`
- **From Python Tuple**: `np.array((1, 2, 3, 4, 5))`
- **Using `fromiter()`**: `np.fromiter((a for a in range(8)), float)`


```python
from random import random

#From Python List
import numpy as np

list=[1,5,6,8]
arr=np.array(list) #list
print(arr)

#From tuple
arr=np.array((1,2,3)) #tupple
print(arr)


```
