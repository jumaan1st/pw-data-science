```python
import threading

import time
from random import random


def calc_square(numbers):
    square = [];
    for number in numbers:
        square.append(number*number)
    time.sleep(1)
    return square


def calc_cube(numbers):
    cube = []
    for number in numbers:
        cube.append(number*number*number)
    time.sleep(1)
    return cube


arr=[];


for i in range(1,11):
    arr.append(i)

print(arr)
start = time.time()
t1=threading.Thread(target=calc_cube,args=(arr,))
t2=threading.Thread(target=calc_square,args=(arr,))

t1.start()
t2.start()

t1.join()
t2.join()
end=time.time()


print(f"Time taken is {end-start} seconds")
print(t1.is_alive())
```
