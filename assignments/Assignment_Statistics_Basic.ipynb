

## **Question 1:*What is the difference between descriptive statistics and inferential statistics? Explain with examples.***

---


**1️⃣ Descriptive Statistics**
Descriptive statistics refers to the branch of statistics that **summarizes, organizes, and presents data** in a meaningful way. It helps in understanding the characteristics of a dataset **without making conclusions beyond it**.

**Example:**

* Marks of five students: `[60, 70, 80, 90, 100]`
* Mean = 80, Median = 80, Standard Deviation ≈ 14
* These measures describe the dataset but **do not provide information about students not included in the data**.

---


**2️⃣ Inferential Statistics**
Inferential statistics refers to the branch of statistics that **uses sample data to make predictions or generalizations about a larger population**.

**Example:**

* If we take the scores of 10 students as a sample, we can use a **t-test** to determine whether the average score of all students in the population is 70.
* Inferential statistics allow **conclusions about the population beyond the observed sample**.




````python
##Example for Descriptive Statistics

import numpy as np

# Sample data
scores = [60, 70, 80, 90, 100]

# Descriptive statistics
mean_score = np.mean(scores)
median_score = np.median(scores)
std_dev = np.std(scores)

print("Descriptive Statistics Example")
print("Scores:", scores)
print("Mean:", mean_score)
print("Median:", median_score)
print("Standard Deviation:", std_dev)
````

````python
##Example for Inferential Statistics

from scipy import stats

# Sample data
sample_scores = [65, 70, 68, 72, 75, 74, 69, 71, 73, 70]

# One-sample t-test (population mean = 70)
t_stat, p_value = stats.ttest_1samp(sample_scores, 70)

print("Inferential Statistics Example")
print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Conclusion: Sample mean is significantly different from 70")
else:
    print("Conclusion: No significant difference from 70")
````

**Definition:**
Sampling is the process of selecting a subset of individuals, items, or observations from a larger population to estimate characteristics of the whole population. It is used to make data collection manageable and to draw conclusions without examining the entire population.

---

## **Types of Sampling: Random vs Stratified**

| Feature        | Random Sampling                                                       | Stratified Sampling                                                                                                                     |
| -------------- | --------------------------------------------------------------------- | --------------------------------------------------------------------------------------------------------------------------------------- |
| **Definition** | Every member of the population has an equal chance of being selected. | Population is divided into subgroups (strata) based on certain characteristics, and samples are taken from each stratum proportionally. |
| **Purpose**    | Simple and unbiased; good when population is homogeneous.             | Ensures representation of all subgroups; useful when population is heterogeneous.                                                       |
| **Complexity** | Simple to implement.                                                  | More complex, requires knowledge of strata.                                                                                             |
| **Bias**       | Low if population is homogeneous.                                     | Low; better representation of subgroups.                                                                                                |
| **Example**    | Picking 50 students randomly from a school.                           | Picking 10 students from each grade in a school to represent all grades.                                                                |

---


Random Sampling


````python
# Randomly select 5 students
random_sample = df.sample(n=5, random_state=42)
print("Random Sample:\n", random_sample)
````



## **Question 3: Define mean, median, and mode. Explain why these measures of centraltendency are important.**

### **1️⃣ Mean (Average)**

**Definition:**
The mean is the sum of all data points divided by the total number of points. It represents the arithmetic average of the dataset.

**Formula:**
[
\text{Mean} = \frac{\sum_{i=1}^{n} x_i}{n}
]

**Example:**
For data ( [5, 8, 12, 20] ):
[
\text{Mean} = \frac{5+8+12+20}{4} = 11.25
]

---

### **2️⃣ Median**

**Definition:**
The median is the middle value of an ordered dataset. If the number of observations is even, it is the average of the two middle numbers.

**Example:**

* Data ( [5, 8, 12, 20, 25] ) → Median = 12 (middle value)
* Data ( [5, 8, 12, 20] ) → Median = ( \frac{8+12}{2} = 10 )

---

### **3️⃣ Mode**

**Definition:**
The mode is the value that occurs most frequently in a dataset. A dataset can have no mode, one mode, or multiple modes.

**Example:**

* Data ( [5, 8, 8, 12, 20] ) → Mode = 8

---

### **Importance of Measures of Central Tendency**

1. **Summarizes Data:** Gives a single value that represents the center of the dataset.
2. **Comparison:** Helps compare different datasets quickly.
3. **Decision Making:** Used in business, economics, and research to make informed decisions.
4. **Foundation for Further Analysis:** Mean, median, and mode are basic inputs for other statistical measures like variance, standard deviation, and skewness.
---

### **Python Code (Jupyter Notebook)**

````python
# Import libraries
import pandas as pd
from scipy import stats

# Sample dataset
data = [5, 8, 12, 20, 8]

# Mean
mean_value = sum(data)/len(data)
print("Mean:", mean_value)

# Median using pandas
median_value = pd.Series(data).median()
print("Median:", median_value)

# Mode using scipy (updated)
mode_result = stats.mode(data, keepdims=True)
mode_value = mode_result.mode[0]
print("Mode:", mode_value)
````



## **Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?**

### **1️⃣ Skewness**

**Definition:**
Skewness measures the **asymmetry** of the data distribution around its mean. It tells us whether the data is **tilted to the left or right**.

**Types of Skewness:**

| Type                           | Description                                                                        | Example                                                              |
| ------------------------------ | ---------------------------------------------------------------------------------- | -------------------------------------------------------------------- |
| **Positive skew (right skew)** | Tail of the distribution is longer on the right. Most data points are on the left. | Income distribution in a country where most earn low, few earn high. |
| **Negative skew (left skew)**  | Tail is longer on the left. Most data points are on the right.                     | Test scores where most students score high, few score very low.      |
| **Zero skew**                  | Symmetrical distribution                                                           | Normal distribution                                                  |

**Implication of Positive Skew:**

* Most of the data points are **less than the mean**.
* The mean is **greater than the median**.
* Presence of **outliers on the higher end**.

---

### **2️⃣ Kurtosis**

**Definition:**
Kurtosis measures the **“tailedness”** or **peakedness** of the data distribution.

**Types of Kurtosis:**

| Type                 | Description                                                      |
| -------------------- | ---------------------------------------------------------------- |
| **Leptokurtic (>3)** | Sharp peak, heavy tails. More outliers than normal distribution. |
| **Mesokurtic (=3)**  | Normal peak, moderate tails. Matches normal distribution.        |
| **Platykurtic (<3)** | Flat peak, light tails. Fewer outliers than normal distribution. |

**Importance:**

* Helps identify **outliers** and the **shape of data**.
* Useful for **risk analysis** in finance and other domains.

---

### **Python Code (Jupyter Notebook)**



````python
# Import libraries
import pandas as pd
from scipy.stats import skew, kurtosis

# Sample dataset
data = [5, 7, 8, 8, 9, 10, 10, 10, 12, 15, 20]

# Skewness
data_skew = skew(data)
print("Skewness:", data_skew)
if data_skew > 0:
    print("Positive skew: Data is right-skewed, mean > median")
elif data_skew < 0:
    print("Negative skew: Data is left-skewed, mean < median")
else:
    print("Symmetrical data: Skewness ~ 0")

# Kurtosis
data_kurtosis = kurtosis(data)  # excess kurtosis; normal distribution = 0
print("Kurtosis (excess):", data_kurtosis)
if data_kurtosis > 0:
    print("Leptokurtic: Sharp peak, heavy tails")
elif data_kurtosis < 0:
    print("Platykurtic: Flat peak, light tails")
else:
    print("Mesokurtic: Moderate peak, like normal distribution")
````

## **numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]**

### **Definition of Measures of Central Tendency**

1. **Mean (Average):**
   The mean is the sum of all numbers divided by the total count. It gives the overall average of the dataset.

   [
   \text{Mean} = \frac{\sum x_i}{n}
   ]

2. **Median:**
   The median is the middle value when the numbers are arranged in order.

   * If the count is **odd**, median is the middle number.
   * If the count is **even**, median is the average of the two middle numbers.

3. **Mode:**
   The mode is the number that appears **most frequently** in the dataset.

   * A dataset may have **no mode**, **one mode**, or **multiple modes**.

---

### **Given Dataset**

```python
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
```

---

### **Python Code (Jupyter Notebook)**


````python
# Import necessary libraries
import pandas as pd
from scipy import stats

# Given dataset
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# 1️⃣ Compute Mean
mean_value = sum(numbers) / len(numbers)
print("Mean:", mean_value)

# 2️⃣ Compute Median
median_value = pd.Series(numbers).median()
print("Median:", median_value)

# 3️⃣ Compute Mode
mode_result = stats.mode(numbers, keepdims=True)
mode_value = mode_result.mode[0]
print("Mode:", mode_value)
````

##list_x = [10, 20, 30, 40, 50]
## list_y = [15, 25, 35, 45, 60]

### **Definitions**

1. **Covariance**
   Covariance measures the **direction of the linear relationship** between two variables.

   * **Positive covariance:** Variables move in the same direction.
   * **Negative covariance:** Variables move in opposite directions.
   * **Formula:**
     [
     \text{Cov}(X, Y) = \frac{\sum (X_i - \bar{X})(Y_i - \bar{Y})}{n-1}
     ]

2. **Correlation Coefficient (Pearson’s r)**
   Correlation measures the **strength and direction** of a linear relationship between two variables.

   * **Value ranges:** -1 to 1

     * 1 → perfect positive correlation
     * -1 → perfect negative correlation
     * 0 → no linear correlation
   * **Formula:**
     [
     r = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y}
     ]

---

### **Given Datasets**

```python
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]
```




````python
# Import libraries
import numpy as np
import pandas as pd

# Given datasets
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# 1️⃣ Compute Covariance
cov_matrix = np.cov(list_x, list_y)  # returns covariance matrix
cov_xy = cov_matrix[0, 1]           # extract covariance between X and Y
print("Covariance between X and Y:", cov_xy)

# 2️⃣ Compute Correlation Coefficient
corr_matrix = np.corrcoef(list_x, list_y)  # returns correlation matrix
corr_xy = corr_matrix[0, 1]               # extract correlation coefficient
print("Correlation coefficient between X and Y:", corr_xy)

# Alternative using pandas
df = pd.DataFrame({'X': list_x, 'Y': list_y})
print("\nCovariance using pandas:\n", df.cov())
print("\nCorrelation using pandas:\n", df.corr())
````

## data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

### **Definition**

**Boxplot (or Whisker Plot):**
A boxplot is a visual representation of the **distribution of a dataset** that shows:

* Minimum value
* First quartile (Q1, 25th percentile)
* Median (Q2, 50th percentile)
* Third quartile (Q3, 75th percentile)
* Maximum value
* Outliers

**Outliers:**
Data points that lie **below Q1 − 1.5×IQR** or **above Q3 + 1.5×IQR**, where **IQR = Q3 − Q1**.

---

### **Given Dataset**

```python
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
```

---

### **Python Code (Jupyter Notebook)**







````python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np

# Given dataset
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# 1️⃣ Draw Boxplot
plt.figure(figsize=(8,5))
plt.boxplot(data, vert=True, patch_artist=True, showmeans=True)
plt.title("Boxplot of Data")
plt.ylabel("Values")
plt.show()

# 2️⃣ Identify Outliers
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower bound:", lower_bound)
print("Upper bound:", upper_bound)
print("Outliers:", outliers)
````

* Explain how you would use covariance and correlation to explore this
relationship.

*  Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]**

### ** Using Covariance and Correlation**

1. **Covariance**

   * Covariance measures the **direction of the relationship** between two variables.
   * **Positive covariance:** When advertising spend increases, daily sales tend to increase.
   * **Negative covariance:** When advertising spend increases, daily sales tend to decrease.
   * Limitation: Covariance **does not indicate the strength** of the relationship because it depends on the scale of variables.

2. **Correlation Coefficient (Pearson’s r)**

   * Measures both **strength and direction** of a linear relationship.
   * **Value range:** -1 to 1

     * 1 → perfect positive linear relationship
     * -1 → perfect negative linear relationship
     * 0 → no linear relationship
   * Advantage: **Standardized**, so easier to interpret than covariance.

**Approach:**

* Compute **covariance** to see if the relationship is positive or negative.
* Compute **correlation coefficient** to understand **how strongly advertising spend affects daily sales**.

---

### **Python Code (Jupyter Notebook)**





````python
# Import libraries
import numpy as np
import pandas as pd

# Given datasets
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# 1️⃣ Covariance
cov_matrix = np.cov(advertising_spend, daily_sales)
cov_xy = cov_matrix[0, 1]
print("Covariance between advertising spend and daily sales:", cov_xy)

# 2️⃣ Correlation Coefficient
corr_matrix = np.corrcoef(advertising_spend, daily_sales)
corr_xy = corr_matrix[0, 1]
print("Correlation coefficient between advertising spend and daily sales:", corr_xy)

# 3️⃣ Alternative using pandas
df = pd.DataFrame({'Advertising_Spend': advertising_spend, 'Daily_Sales': daily_sales})
print("\nCovariance using pandas:\n", df.cov())
print("\nCorrelation using pandas:\n", df.corr())
````

* Explain which **summary statistics and visualizations** (e.g., mean, standard deviation, histogram) you’d use.
* Write **Python code** to create a **histogram using Matplotlib** for the survey data:

```python
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]
```

---


To understand the **distribution of customer satisfaction scores**, we can use:

#### **1️⃣ Summary Statistics**

* **Mean:** The average score → indicates overall satisfaction.
* **Median:** Middle value → identifies central tendency and is robust to outliers.
* **Mode:** Most frequent score → shows the most common customer opinion.
* **Standard Deviation (SD):** Measures variability → higher SD indicates more spread.
* **Minimum & Maximum:** Shows the range of scores.

#### **2️⃣ Visualizations**

* **Histogram:** Displays the frequency of each score → shows distribution and skewness.
* **Boxplot (optional):** Shows median, quartiles, and potential outliers.

**Purpose:**

* Summary statistics provide **numerical understanding** of central tendency and spread.
* Histograms allow the team to **visualize how scores are distributed**, which is critical for product launch planning.

---

### **Python Code (Jupyter Notebook)**



````python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd

# Survey data
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# 1️⃣ Compute Summary Statistics
mean_score = pd.Series(survey_scores).mean()
median_score = pd.Series(survey_scores).median()
mode_score = pd.Series(survey_scores).mode()[0]
std_dev = pd.Series(survey_scores).std()

print("Mean:", mean_score)
print("Median:", median_score)
print("Mode:", mode_score)
print("Standard Deviation:", std_dev)

# 2️⃣ Create Histogram
plt.figure(figsize=(8,5))
plt.hist(survey_scores, bins=7, color='skyblue', edgecolor='black')
plt.title("Histogram of Customer Satisfaction Scores")
plt.xlabel("Survey Score")
plt.ylabel("Frequency")
plt.xticks(range(1, 11))  # Scores range from 1 to 10
plt.show()
````
