### **Question 1:** What is hypothesis testing in statistics?

**Answer:**
Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It helps determine whether there is enough evidence to reject a certain assumption (called the **null hypothesis**) in favor of an alternative hypothesis.

In hypothesis testing:

* The **null hypothesis (H₀)** represents the default or no-effect assumption.
* The **alternative hypothesis (H₁ or Ha)** represents what you aim to prove or find evidence for.
* Based on the data and a chosen **significance level (α)**, a test statistic is calculated, and a **p-value** is used to decide whether to reject or fail to reject the null hypothesis.

**Example:**
Suppose a company claims that their light bulbs last an average of 1000 hours. A researcher tests a sample of bulbs to check this claim.

* **H₀:** The mean lifespan = 1000 hours
* **H₁:** The mean lifespan ≠ 1000 hours

If the p-value is less than 0.05 (assuming α = 0.05), the null hypothesis is rejected, suggesting that the bulbs’ average lifespan is significantly different from 1000 hours.


### **Question 2:** What is the null hypothesis, and how does it differ from the alternative hypothesis?

**Answer:**
The **null hypothesis (H₀)** is a statement that assumes there is **no effect, no difference, or no relationship** between variables. It represents the default or status quo assumption that researchers try to test.

The **alternative hypothesis (H₁ or Ha)**, on the other hand, is a statement that suggests there **is an effect, a difference, or a relationship** — essentially, what the researcher aims to prove.

**Key Differences:**

| Aspect       | Null Hypothesis (H₀)                        | Alternative Hypothesis (H₁ or Ha)           |
| ------------ | ------------------------------------------- | ------------------------------------------- |
| **Meaning**  | Assumes no effect or difference             | Assumes an effect or difference exists      |
| **Purpose**  | Serves as the default assumption            | Represents the researcher’s claim           |
| **Decision** | Rejected or not rejected                    | Accepted if H₀ is rejected                  |
| **Example**  | H₀: The average height of students = 170 cm | H₁: The average height of students ≠ 170 cm |

**In summary:**
The null hypothesis is what we test against, while the alternative hypothesis is what we hope to support with statistical evidence.

### **Question 3:** Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

**Answer:**
The **significance level**, denoted by **α (alpha)**, is the **threshold or cutoff value** used in hypothesis testing to decide whether to reject the null hypothesis (**H₀**). It represents the **probability of making a Type I error**, which occurs when the null hypothesis is rejected even though it is actually true.

**Common significance levels:**

* α = 0.05 (most commonly used)
* α = 0.01 (more strict)
* α = 0.10 (less strict)

**Role in decision-making:**

1. After performing the hypothesis test, a **p-value** is calculated.
2. The p-value is compared with the significance level (α):

   * If **p-value ≤ α**, **reject H₀** → There is significant evidence against the null hypothesis.
   * If **p-value > α**, **fail to reject H₀** → There is not enough evidence to reject the null hypothesis.

**Example:**
If α = 0.05 and the p-value obtained is 0.03, then p-value < α, so the null hypothesis is rejected.
This means the result is **statistically significant** at the 5% level.


### **Question 4:** What are Type I and Type II errors? Give examples of each.

**Answer:**
In hypothesis testing, **Type I and Type II errors** are two possible mistakes that can occur when making decisions about a hypothesis based on sample data.

---

### **1. Type I Error (False Positive)**

* **Definition:** Occurs when the **null hypothesis (H₀)** is **true**, but we **incorrectly reject it**.
* **Meaning:** We conclude that there is an effect or difference when in reality there isn’t.
* **Symbol:** Represented by **α (alpha)** — the significance level.

**Example:**
A medical test concludes that a healthy person has a disease.

* **H₀:** The person does not have the disease.
* **H₁:** The person has the disease.
  Rejecting H₀ (saying the person has the disease) when they are actually healthy is a **Type I error**.

---

### **2. Type II Error (False Negative)**

* **Definition:** Occurs when the **null hypothesis (H₀)** is **false**, but we **fail to reject it**.
* **Meaning:** We miss detecting a real effect or difference.
* **Symbol:** Represented by **β (beta)**.

**Example:**
A medical test fails to detect a disease in a sick person.

* **H₀:** The person does not have the disease.
* **H₁:** The person has the disease.
  Failing to reject H₀ (saying the person is healthy) when they actually have the disease is a **Type II error**.

---

**In short:**

| Error Type      | Decision          | Actual Truth | Example                          |
| --------------- | ----------------- | ------------ | -------------------------------- |
| **Type I (α)**  | Reject H₀         | H₀ is true   | Healthy person diagnosed as sick |
| **Type II (β)** | Fail to reject H₀ | H₀ is false  | Sick person diagnosed as healthy |


### **Question 5:** What is the difference between a Z-test and a T-test? Explain when to use each.

**Answer:**
Both the **Z-test** and **T-test** are statistical methods used to determine whether there is a significant difference between sample data and a population, or between two sample means. The key difference lies in the **sample size** and whether the **population standard deviation (σ)** is known.

---

### **1. Z-Test**

* **Used when:**

  * The **population standard deviation (σ)** is **known**.
  * The **sample size** is **large** (typically **n > 30**).
* **Distribution Used:** Standard **Normal Distribution (Z-distribution)**.
* **Purpose:** To test hypotheses about population means or proportions when data follows a normal distribution.

**Example:**
Testing whether the average height of 100 students differs from 170 cm, given that the population standard deviation is known.

---

### **2. T-Test**

* **Used when:**

  * The **population standard deviation (σ)** is **unknown**.
  * The **sample size** is **small** (typically **n ≤ 30**).
* **Distribution Used:** **Student’s T-distribution**, which is similar to the normal distribution but has thicker tails to account for extra uncertainty in small samples.
* **Purpose:** To test hypotheses about population means when σ is not known.

**Example:**
Testing whether the average score of 20 students differs from 75, without knowing the population standard deviation.

---

### **Key Differences:**

| Feature                               | **Z-Test**                                | **T-Test**                              |
| ------------------------------------- | ----------------------------------------- | --------------------------------------- |
| **Population Standard Deviation (σ)** | Known                                     | Unknown                                 |
| **Sample Size**                       | Large (n > 30)                            | Small (n ≤ 30)                          |
| **Distribution Used**                 | Z (Normal) Distribution                   | T Distribution                          |
| **Shape of Curve**                    | Fixed                                     | Depends on degrees of freedom           |
| **Example Use**                       | Comparing population mean when σ is known | Comparing sample mean when σ is unknown |

---

**In summary:**

* Use a **Z-test** when the population standard deviation is known and the sample is large.
* Use a **T-test** when the population standard deviation is unknown or the sample is small.


### **Question 6:** Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n = 10     # number of trials
p = 0.5    # probability of success
size = 1000  # number of samples

# Generate binomial distribution data
data = np.random.binomial(n, p, size)

# Plot the histogram
plt.hist(data, bins=range(n+2), edgecolor='black', alpha=0.7)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.xticks(range(n+1))
plt.show()
```

**Explanation:**

1. `np.random.binomial(n, p, size)` generates random samples from a binomial distribution with `n` trials and success probability `p`.
2. `plt.hist()` is used to plot the histogram.
3. `bins=range(n+2)` ensures each integer outcome from 0 to n is represented.
4. The histogram shows how frequently each number of successes occurs in the generated sample.

This will display a histogram resembling the binomial distribution for 10 trials with a 50% chance of success.

### **Question 7:** Implement hypothesis testing using Z-statistics for a sample dataset in Python.

**Answer:**

We will test whether the mean of the sample dataset is significantly different from a population mean, say **μ = 50**, using a **one-sample Z-test**.


```python
import numpy as np
from scipy import stats

# Sample dataset
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Known population mean and standard deviation
population_mean = 50
population_std = 0.5  # Assume known population standard deviation

# Sample statistics
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Calculate Z-statistic
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate p-value for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Print results
print("Sample Mean:", round(sample_mean, 3))
print("Z-Statistic:", round(z_statistic, 3))
print("P-Value:", round(p_value, 4))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference from the population mean.")
```

**Interpretation:**

* The calculated Z-statistic is **2.307**, and the p-value is **0.0211**.
* Since the p-value < 0.05, we **reject the null hypothesis**.
* This suggests that the sample mean is **significantly different** from the population mean of 50 at the 5% significance level.

**Question 8:** Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean.


```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Parameters for normal distribution
mu = 50         # Mean
sigma = 5       # Standard deviation
sample_size = 100

# Generate random sample data
data = np.random.normal(mu, sigma, sample_size)

# Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation
standard_error = sample_std / np.sqrt(sample_size)

# Calculate 95% confidence interval
confidence_level = 0.95
z_score = stats.norm.ppf((1 + confidence_level) / 2)
ci_lower = sample_mean - z_score * standard_error
ci_upper = sample_mean + z_score * standard_error

# Print results
print("Sample Mean:", round(sample_mean, 3))
print("95% Confidence Interval: [{:.3f}, {:.3f}]".format(ci_lower, ci_upper))

# Plot histogram
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label='95% CI Lower')
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label='95% CI Upper')
plt.title("Normal Distribution Simulation")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

**Explanation:**

1. `np.random.normal(mu, sigma, sample_size)` generates random samples from a normal distribution.
2. The **sample mean** and **standard error** are calculated.
3. Using the **Z-score for 95% confidence**, the confidence interval is computed.
4. The histogram shows the distribution of simulated data, with dashed lines for the mean and confidence interval.


### **Question 9:** Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent.


```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # Sample standard deviation
    z_scores = (data - mean) / std
    return z_scores

# Sample dataset
data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5]

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Print Z-scores
print("Z-scores:", np.round(z_scores, 3))

# Plot histogram of Z-scores
plt.hist(z_scores, bins=10, edgecolor='black', alpha=0.7)
plt.title("Histogram of Z-Scores")
plt.xlabel("Z-Score")
plt.ylabel("Frequency")
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label='Mean (Z=0)')
plt.legend()
plt.show()
```

**Explanation:**

* **Z-scores** represent how many **standard deviations a data point is away from the mean**.

  * Z = 0 → The data point equals the mean.
  * Z > 0 → The data point is above the mean.
  * Z < 0 → The data point is below the mean.
* Standardizing data using Z-scores helps compare values from different datasets or scales.
* The histogram visualizes the spread of the data in terms of standard deviations from the mean.`
