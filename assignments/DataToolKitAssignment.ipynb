### 1. **What is NumPy, and why is it widely used in Python?**
   **Answer**: NumPy is a Python library for numerical computing, providing support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays efficiently. It is widely used because it offers fast, vectorized operations, supports multi-dimensional data, and integrates well with other scientific libraries like Pandas, SciPy, and Matplotlib. Its efficiency comes from optimized C-based implementations, making it ideal for data science, machine learning, and scientific computations.

```python
import numpy as np
arr = np.array([1, 2, 3, 4])
print(arr * 2)
```

### 2. **How does broadcasting work in NumPy?**
   **Answer**: Broadcasting allows NumPy to perform operations on arrays of different shapes by automatically expanding their dimensions to match. It follows rules where smaller arrays are "stretched" to match the shape of larger arrays without copying data, enabling efficient computation. Arrays must be compatible (same dimensions or one dimension is 1).

```python
import numpy as np
a = np.array([[1, 2], [3, 4]])
b = np.array([10, 20])
result = a + b
print(result)
```

### 3. **What is a Pandas DataFrame?**
   **Answer**: A Pandas DataFrame is a 2-dimensional, labeled data structure with columns of potentially different data types, similar to a spreadsheet or SQL table. It is used for data manipulation, analysis, and cleaning, offering flexible indexing and powerful methods for filtering, grouping, and merging data.

```python
import pandas as pd
data = {'Name': ['Alice', 'Bob'], 'Age': [25, 30]}
df = pd.DataFrame(data)
print(df)
```

### 4. **Explain the use of the groupby() method in Pandas.**
   **Answer**: The `groupby()` method in Pandas groups data by one or more columns and applies an aggregation function (e.g., mean, sum, count) to each group. It is used for summarizing data, analyzing trends, or computing group-level statistics.

```python
import pandas as pd
df = pd.DataFrame({'City': ['Bangalore', 'Mysore', 'Bangalore', 'Mysore'], 'Sales': [200, 300, 150, 400]})
grouped = df.groupby('City').sum()
print(grouped)
```

### 10. **What is the significance of hierarchical indexing in Pandas?**
 **Answer**: Hierarchical indexing (MultiIndex) in Pandas allows organizing data with multiple levels of indices, enabling complex data analysis across multiple dimensions. It simplifies grouping, slicing, and reshaping data for tasks like time series or panel data analysis.

```python
import pandas as pd
arrays = [['A', 'A', 'B', 'B'], [1, 2, 1, 2]]
index = pd.MultiIndex.from_arrays(arrays, names=('Group', 'Number'))
df = pd.DataFrame({'Value': [10, 20, 30, 40]}, index=index)
print(df)
```

### 17. **Explain the difference between apply() and map() in Pandas.**
**Answer**: In Pandas, `apply()` is used to apply a function along an axis of a DataFrame or to a Series, allowing complex operations. `map()` is used only on a Series to apply a function or mapping dictionary to each element. `apply()` is more flexible but slower than `map()`.`

```python
import pandas as pd
s = pd.Series([1, 2, 3])
print(s.map(lambda x: x * 2))
df = pd.DataFrame({'A': [1, 2, 3]})
print(df.apply(lambda x: x * 2))
```

### 19. **How does Pandas simplify time series analysis?**
**Answer**: Pandas simplifies time series analysis with datetime indexing, time-based slicing, resampling, rolling windows, and time zone handling. It provides functions like `to_datetime()`, `resample()`, and `rolling()` for analyzing temporal data.

```python
import pandas as pd
dates = pd.date_range('2023-01-01', periods=3, freq='D')
df = pd.DataFrame({'Value': [10, 20, 30]}, index=dates)
print(df.resample('2D').sum())
```

### 20. **What is the role of a pivot table in Pandas?**
**Answer**: A pivot table in Pandas reshapes data by aggregating values based on specified index, columns, and aggregation functions (e.g., mean, sum). It is used for summarizing and analyzing data, similar to Excel pivot tables.


```python
import pandas as pd
df = pd.DataFrame({'A': ['foo', 'foo', 'bar'], 'B': ['x', 'y', 'x'], 'C': [1, 2, 3]})
pivot = df.pivot_table(values='C', index='A', columns='B', aggfunc='sum')
print(pivot)
```

### 21. **Why is NumPy’s array slicing faster than Python’s list slicing?**
**Answer**: NumPy’s array slicing is faster because arrays are stored contiguously in memory with fixed data types, enabling optimized C-based operations. Python lists are heterogeneous, stored as pointers to objects, requiring slower memory access and iteration.

```python
import numpy as np
arr = np.array([1, 2, 3, 4, 5])
print(arr[1:4])  #fast slicing
lst = [1, 2, 3, 4, 5]
print(lst[1:4])  #Slow slicing
```
